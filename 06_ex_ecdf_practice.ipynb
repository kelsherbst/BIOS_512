# ECDF plots

```r
# uncomment the following:

library('tidyverse')
options(repr.plot.width = 4, repr.plot.height = 3, repr.plot.res = 300)
```

This cell provides the data ðŸ‘‡.

```r
otter.data = read.csv('https://jcoliver.github.io/learn-r/data/otter-mandible-data.csv') %>%
    gather(characteristic, value, -species, -museum, -accession) %>%
     drop_na() %>%
     group_by(species, characteristic) %>%
     arrange(value) %>%
     mutate(ecdf_y = row_number() / n()) %>%
     ungroup()

 otter.data %>% head
```

Make an ECDF plot of `value` and:
- facet by characteristic
- allow the x-axis of each facetted plot to be independent (for the value of the the `scales` argument in `facet_wrap` use `"free_x"`)
- color/group by species  

Species names are typically italicized. Use `theme` and `element_text` to adjust the legend text so that it is italic.

Which otter species has the smallest skull?

#

```r
options(repr.plot.width=8, repr.plot.height=5.5)
p = ggplot(otter.data, aes(x = value, y=ecdf_y, color = species)) + geom_point(size=0.2) +
    geom_step(direction='hv') + facet_wrap(~characteristic, scales="free_x") + #free_x lets each axis scale independently
    scale_y_continuous(labels = scales::percent_format()) + # scales::percent_format changes to percentage by defaul (multiplied by 100)
    scale_color_brewer(palette = 'Dark2') +
    labs(y = 'Percentile') + theme_gray()
p
    
```

## ECDF using stat_ecdf

```r
options(repr.plot.width=8, repr.plot.height=5.5)
p = ggplot(otter.data, aes(x = value, y=ecdf_y, color = species)) + 
    stat_ecdf(geom = "step") + 
    facet_wrap(~characteristic, scales="free_x") + #free_x lets each axis scale independently
    scale_y_continuous(labels = scales::percent_format()) + # scales::percent_format changes to percentage by defaul (multiplied by 100)
    scale_color_brewer(palette = 'Dark2') +
    labs(y = 'Percentile') + theme_gray()
p
```
